````python
#1
import pandas as pd

# Load in the dataset
df = pd.read_csv(r"C:\Users\Brian\.vscode\visualization\assignment\data\college_completion.csv")


````

````python
#2
# Check dimensions
print(f"Dimensions of dataset: {df.shape}")

#Check variable names
print(f"Variables included: {df.columns.tolist()}")

# Show first few rows
print(df.head())
````

````python
#3
# Cross tabulate control and level
crosstab = pd.crosstab(df['control'], df['level'])
print(crosstab)
````

This table shows the number of institutions by their control type (Private for-profit, Private not-for-profit, and Public) and level (2-year and 4-year). Public are the most prevalent in 2-year while Private NFP is the least prevalent. However, Private NFP seems to be the most prevalent for 4-year education while Private FP is the least prevalent.
````python
#4
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram
sns.histplot(df['grad_100_value'], bins=30, kde=True)
plt.title("Histogram of Graduation Rate (grad_100_value)")
plt.show()

# Kernel Density Plot
sns.kdeplot(df['grad_100_value'], fill=True)
plt.title("KDE of Graduation Rate (grad_100_value)")
plt.show()

# Boxplot
sns.boxplot(x=df['grad_100_value'])
plt.title("Boxplot of Graduation Rate")
plt.show()

# Statistical description
print(df['grad_100_value'].describe())
````

````python
#5
# KDE by CONTROL
sns.kdeplot(data=df, x='grad_100_value', hue='control', fill=True)
plt.title("Graduation Rate by Control Type")
plt.show()

# KDE by LEVEL
sns.kdeplot(data=df, x='grad_100_value', hue='level', fill=True)
plt.title("Graduation Rate by Level (2-year vs 4-year)")
plt.show()

# Grouped statistical descriptions
print(df.groupby(['level', 'control'])['grad_100_value'].describe())

````

4-year institutions seem to have generally higher graduation rates than 2-year institutions. At the same time, Private NFP institutions appear to show the highest graduation rates.

````python
#6
df['levelXcontrol'] = df['level'] + ', ' + df['control']

# KDE plot of grad_100_value by level + control interaction
sns.kdeplot(data=df, x='grad_100_value', hue='levelXcontrol', fill=True)
plt.title("Graduation Rate by Institution Type")
plt.show()
````

4-year Private NFP shows a tendency highest graduation rates while Public 2-year institutions are the opposite.

````python
#8
# KDE plot of aid_value
sns.kdeplot(df['aid_value'], fill=True)
plt.title("Kernel Density Plot of Aid Value")
plt.show()

# Grouped KDE plot by level and control
sns.kdeplot(data=df, x='aid_value', hue='levelXcontrol', fill=True)
plt.title("Aid Distribution by Institution Type")
plt.show()

# Statistical description of aid_value by level and control
print(df.groupby(['level', 'control'])['aid_value'].describe())
````

Aid distibutions vary significantly 4 year Private NFP schools and the others. Private NFP 4-yeay institutions generally provide more aid compared to the other controls. They also have the most variation though.

````python
#basic scatterplot
sns.scatterplot(data=df, x='aid_value', y='grad_100_value')
plt.title("Graduation Rate vs. Aid Value")
plt.show()

# Grouped by LEVEL
sns.scatterplot(data=df, x='aid_value', y='grad_100_value', hue='level')
plt.title("Graduation Rate vs. Aid Value (Grouped by Level)")
plt.show()

# Grouped by CONTROL
sns.scatterplot(data=df, x='aid_value', y='grad_100_value', hue='control')
plt.title("Graduation Rate vs. Aid Value (Grouped by Control)")
plt.show()
````

There is a trend showing that higher financual aid correlates with higher graduation rates. This effect is more prominent in 4-year institutions, especially the Private NFP ones. Public and Private FP seem to have less variation than Private NFP.

3. The **hazard rate** gives the instantaneous risk of an event occurring, given that it hasn't already happened. It is defined as
$$
h(t) = \dfrac{f(t)}{1-F(t)}
$$
where $F(t)$ is the distribution of arrival times and $f(t)=F'(t)$ is the density of arrival times. Roughly, conditional on $T \ge t$, what is the probability that $T=t$? The cumulative hazard is given by
$$
H(t) = \int_0^t h(z) dz.
$$
The following function computes the hazard rate and cumulative hazard using something called the Nelson-Ahlen estimator. Run your data through it. Interpret the plots for the hazard rate and cumulative hazard rate. What patterns do you see?

```

    hazard_rate = []
    times = []
    for t in arrivals:
        select = (df['survival']==t)*(df['Overall Survival Status']=='Deceased')
        deaths_t = df[select].shape[0]
        if deaths_t > 0:
            hazard_rate.append(deaths_t/at_risk)
            times.append(t)
```
